```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G1 = nx.DiGraph()
G1.add_nodes_from([i for i in range(6)])
G1.add_weighted_edges_from((
    (0, 1, 16), (0, 2, 13),
    (1, 3, 12), 
    (2, 1, 4), (2, 4, 14),
    (3, 5, 20),
    (4, 3, 7), (4, 5, 4)
))
pos1 = {
    0: (-0.5, 0),
    1: (0, 0.2),
    2: (0, -0.2),
    3: (0.5, 0.2),
    4: (0.5, -0.2),
    5: (1, 0)
}
edge_labels1 = nx.get_edge_attributes(G1, 'weight')
nx.draw(G1, pos1, with_labels=True, node_size=1000)
nx.draw_networkx_edge_labels(G1, pos1, edge_labels=edge_labels1)
```
