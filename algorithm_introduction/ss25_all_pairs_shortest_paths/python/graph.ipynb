```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G1 = nx.DiGraph()
G1.add_nodes_from([i for i in range(5)])
G1.add_weighted_edges_from((
    (0, 1, 3), (0, 2, 8), (0, 4, -4),
    (1, 3, 1), (1, 4, 7),
    (2, 1, 4),
    (3, 0, 2), (3, 2, -5),
    (4, 3, 6)
))
pos1 = {
    0: (-0.35, 0.65),
    1: (0, 1),
    2: (0.35, 0.65),
    3: (0.2, 0),
    4: (-0.2, 0)
}
labels1 = nx.get_edge_attributes(G1, 'weight')
nx.draw(G1, pos1, with_labels=True, node_size=1000)
nx.draw_networkx_edge_labels(G1, pos1, edge_labels=labels1, font_size=20, label_pos=0.25)
```
