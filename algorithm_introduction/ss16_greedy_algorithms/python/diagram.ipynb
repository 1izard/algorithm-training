```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

```python
G = nx.Graph()

G.add_edge('a', 'b', weight=6)
G.add_edge('a', 'c', weight=15)
G.add_edge('b', 'd', weight=10)
G.add_edge('d', 'c', weight=1)
G.add_edge('d', 'e', weight=5)
G.add_edge('c', 'e', weight=4)
G.add_edge('d', 'f', weight=9)
G.add_edge('e', 'h', weight=3)
G.add_edge('f', 'g', weight=14)
G.add_edge('f', 'h', weight=8)
G.add_edge('g', 'h', weight=12)
G.add_edge('e', 'i', weight=7)
G.add_edge('g', 'i', weight=13)

pos = nx.spring_layout(G)
edge_labels = dict([((u, v), d['weight']) for u, v, d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw_networkx_labels(G, pos)
nx.draw(G, pos)
```
