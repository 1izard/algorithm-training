```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()
G.add_nodes_from([i for i in range(5)])
G.add_weighted_edges_from((
    (0, 1, 6), (0, 3, 7),
    (1, 2, 5), (1, 3, 8), (1, 4, -4),
    (2, 1, -2),
    (3, 2, -3), (3, 4, 9),
    (4, 0, 2), (4, 2, 7)
))
pos = {
    0: (-0.5, 0),
    1: (0, 0.5),
    2: (0.5, 0.5),
    3: (0, -0.5),
    4: (0.5, -0.5)
}
labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_size=1000)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, label_pos=0.4, font_size=20)
```
