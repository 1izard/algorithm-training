```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G1 = nx.Graph()
G1.add_nodes_from([i for i in range(7)])
G1.add_edges_from([
    (0, 1),
    (1, 2),
    (2, 3),
    (2, 4),
    (3, 4),
    (3, 5),
    (4, 5),
    (3, 6)
])
pos1 = {
    0: (0, -1),
    1: (0, 0),
    2: (1, 0),
    3: (2, 0),
    4: (1, -1),
    5: (2, -1),
    6: (3, -1)
}
nx.draw(G1, pos1, with_labels=True, node_size=1000)
nx.draw_networkx_edges(G1, pos1)
```
